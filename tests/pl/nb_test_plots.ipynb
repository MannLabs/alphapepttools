```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt

from alphatools.pl.figure import create_figure
from alphatools.pl.plots import label_plot

# Shared data
x = [2, 1, 2, 1, 2, 1]
y = [2, 2, 3, 3, 1, 1]
labels = ["middle_right", "middle_left", "top_right", "top_left", "bottom_right", "bottom_left"]

A_DISPLAY_START = 3.20
Y_PADDING_FACTOR = 10

# Create figure with 1 row, 2 cols
fig, axm = create_figure(1, 2, figsize=(6, 3))
fig.canvas.draw()

# First subplot: anchors=None
ax1 = axm.next()
ax1.scatter(x, y, color="blue", zorder=3)
label_plot(
    ax=ax1,
    x_values=x,
    y_values=y,
    labels=labels,
    x_anchors=None,
    y_display_start=A_DISPLAY_START,
    y_padding_factor=Y_PADDING_FACTOR,
    line_operation="add_return",
)
ax1.set_title("anchors=None")

# Second subplot: anchors=(0.5, 2.5)
ax2 = axm.next()
ax2.scatter(x, y, color="blue", zorder=3)
lins = label_plot(
    ax=ax2,
    x_values=x,
    y_values=y,
    labels=labels,
    x_anchors=(0.5, 2.5),
    y_display_start=A_DISPLAY_START,
    y_padding_factor=Y_PADDING_FACTOR,
    line_operation="add_return",
)
ax2.set_title("anchors=(0.5, 2.5)")


plt.show()

print("Label positioning lines int he second subplot:")
print(lins)
```
